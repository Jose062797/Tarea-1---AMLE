### Pregunta 1.8

El df contiene variables con missing values (`NaN`). Impute los `NaN` con el método que estime conveniente, justificando su decisión.

¿Es pertinente eliminar alguna de estas variables? Hágalo si es el caso.

```python
# Calculamos el porcentaje de valores nulos por columna
missing_percentage = df.isna().mean().sort_values(ascending=False) * 100

print("Porcentaje de valores nulos por columna:")
print(missing_percentage[missing_percentage > 0])

# Visualizamos los valores faltantes
import missingno as msno
plt.figure(figsize=(12, 6))
msno.matrix(df)
plt.title('Matriz de valores faltantes')
plt.show()
```

```python
# Examinamos la distribución de algunas variables para justificar el método de imputación
plt.figure(figsize=(14, 5))
```

Tras analizar el porcentaje de valores faltantes por variable, se observa que ninguna supera el umbral del 10%, por lo que decidimos conservar todas las columnas. La variable con más datos ausentes es "Net migration" con aproximadamente 9% de valores faltantes.

Para la imputación de valores faltantes, utilizamos un enfoque diferenciado:

1. Para variables con distribución notablemente sesgada (GDP per capita y Pop. Density), utilizamos la mediana como método de imputación, ya que es más robusta frente a valores extremos y representa mejor la tendencia central en distribuciones asimétricas.

2. Para el resto de variables numéricas, utilizamos la media aritmética, que proporciona una estimación adecuada cuando la distribución es más simétrica.

Este enfoque de imputación nos permite mantener la integridad del dataset preservando la dimensionalidad original y las relaciones entre variables, sin introducir sesgos significativos, lo cual es fundamental para nuestro análisis exploratorio posterior.

```python
# Para variables numéricas con distribución sesgada, usamos la mediana
skewed_columns = ['GDP ($ per capita)', 'Pop. Density (per sq. mi.)']
for col in skewed_columns:
    if col in df.columns and df[col].isna().sum() > 0:
        df[col].fillna(df[col].median(), inplace=True)

# Para el resto de variables numéricas, usamos la media
df.fillna(df.mean(numeric_only=True), inplace=True)

# Verificamos que no queden valores nulos
print("Valores nulos restantes:", df.isna().sum().sum())
```

Analizar la correlación entre variables es fundamental para comprender las relaciones lineales en el dataset y orientar la construcción de modelos explicativos del PIB per cápita.

Las correlaciones pueden clasificarse en tres grupos principales:

1. Correlaciones fuertes positivas (r > 0.5):

   - 'Service' (0.53): La mayor participación del sector servicios refleja la transición hacia economías post-industriales características de países desarrollados.
   - 'Literacy (%)' (0.50): Confirma la teoría del capital humano, donde mayor educación se traduce en mayor productividad y crecimiento económico.

2. Correlaciones fuertes negativas (r < -0.5):
   - 'Birthrate' (-0.65): Consistente con la teoría de la transición demográfica, donde sociedades más prósperas tienden a reducir sus tasas de natalidad.
   - 'Infant mortality' (-0.60): Un claro indicador inverso de desarrollo sanitario y bienestar.
   - 'Agriculture (%)' (-0.57): Refleja la ley de Engel y las etapas de desarrollo económico de Rostow, donde economías más avanzadas reducen su dependencia del sector primario.

3. Correlaciones débiles (|r| < 0.3):
   - Variables como 'Area', 'Coastline' y 'Population' muestran correlaciones cercanas a cero, sugiriendo que el tamaño geográfico o demográfico per se no determina la prosperidad económica.

Este análisis es crucial por múltiples razones:

1. Identifica potenciales predictores para modelos explicativos, guiando la selección de variables.
2. Detecta posible multicolinealidad, lo que afectaría la interpretabilidad de coeficientes en regresiones múltiples.
3. Permite contrastar empíricamente hipótesis derivadas de teorías económicas sobre desarrollo y crecimiento.
4. Revela posibles mecanismos de transmisión entre factores socioeconómicos y prosperidad.

Sin embargo, debemos recordar que correlación no implica causalidad, y algunas relaciones podrían ser no lineales o estar mediadas por terceras variables no observadas.

### Pregunta 1.14

Detecte las observaciones outliers de las tres variables seleccionadas en la pregunta anterior. Además, impute estas observaciones si usted lo considera necesario. Justifique su decisión.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Recalculamos las correlaciones por si no están en memoria
correlations = df.corr(numeric_only=True)['GDP ($ per capita)'].drop('GDP ($ per capita)')
top_3_gdp_total = correlations.abs().sort_values(ascending=False).head(3).index.tolist()

# Creamos un diccionario para guardar los outliers
outliers_dict = {}

# Creamos figuras para visualizar antes y después
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))
fig.suptitle('Outliers antes y después de la imputación (mediana)', fontsize=14)

for i, var in enumerate(top_3_gdp_total):
    # Detectar outliers
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    median = df[var].median()

    # Guardar índices de outliers
    outliers_dict[var] = df[(df[var] < lower) | (df[var] > upper)].index.tolist()

    # Boxplot antes
    sns.boxplot(data=df, x=var, ax=axes[i, 0], color='tomato')
    axes[i, 0].set_title(f'{var} (antes)')

    # Imputar con mediana
    df.loc[(df[var] < lower) | (df[var] > upper), var] = median

    # Boxplot después
    sns.boxplot(data=df, x=var, ax=axes[i, 1], color='seagreen')
    axes[i, 1].set_title(f'{var} (después)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Se detectaron valores atípicos (*outliers*) en las variables más correlacionadas con el PIB total: `GDP totalpais`, `Population` y `Area (sq. mi.)`, utilizando el método del rango intercuartílico (IQR). Este método define como outliers a las observaciones que se ubican fuera del rango [Q1 – 1.5×IQR, Q3 + 1.5×IQR], lo cual es estándar en análisis exploratorios.

Dado que estos valores extremos pueden sesgar tanto la interpretación como los resultados de modelos estadísticos (por ejemplo, afectar la media, la varianza o la pendiente de una regresión), se optó por **imputarlos utilizando la mediana** de cada variable. La mediana es una medida robusta que no se ve afectada por valores extremos y permite conservar la estructura general de los datos sin eliminar observaciones.

Visualmente, los boxplots muestran cómo se reducen las asimetrías y valores extremos tras la imputación. Si bien esto implica una ligera pérdida de variabilidad, mejora la estabilidad del análisis posterior, especialmente en contextos donde se asumirá normalidad o se aplicarán modelos sensibles a *outliers*.

¿Cuál fue la fuente más contaminante en la mayoría de países en 2007?


```python
# Filtrar los datos para el año 2007
df_co2_2007 = df_co2[df_co2['Year'].dt.year == 2007]

# Seleccionar solo las columnas de interés (excluyendo Per Capita y Bunker fuels)
fuentes_emision = ['Solid Fuel', 'Liquid Fuel', 'Gas Fuel', 'Cement', 'Gas Flaring']

# Crear un DataFrame para almacenar los rankings
rankings = pd.DataFrame(index=df_co2_2007['Country'], columns=fuentes_emision)

# Para cada país, calcular el ranking de las fuentes
for pais in df_co2_2007['Country'].unique():
    # Obtener los datos del país
    datos_pais = df_co2_2007[df_co2_2007['Country'] == pais][fuentes_emision].iloc[0]

    # Eliminar fuentes con valor NaN (no reportadas)
    datos_validos = datos_pais.dropna()

    if len(datos_validos) > 0:  # Si hay al menos una fuente reportada
        # Crear el ranking (1 es la mayor fuente, valores más altos son menos contaminantes)
        ranking_pais = datos_validos.rank(method='min', ascending=False)

        # Asignar el ranking al DataFrame
        for fuente in ranking_pais.index:
            rankings.at[pais, fuente] = int(ranking_pais[fuente])

# Eliminar filas con todos NaN (países sin datos de emisiones por fuente)
rankings = rankings.dropna(how='all')

# Crear una figura para los histogramas
plt.figure(figsize=(15, 10))

# Para cada fuente, crear un histograma de su posición en los rankings
for i, fuente in enumerate(fuentes_emision):
    plt.subplot(2, 3, i+1)

    # Contar frecuencias de cada posición en el ranking para esta fuente
    conteo = rankings[fuente].value_counts().sort_index()

    # Crear el histograma
    plt.bar(conteo.index, conteo.values, color=f'C{i}')

    # Añadir etiquetas y título
    plt.title(f'Ranking de {fuente}')
    plt.xlabel('Posición en el ranking (1=mayor emisión)')
    plt.ylabel('Número de países')
    plt.xticks(range(1, 6))

    # Añadir etiquetas con el número exacto sobre cada barra
    for j, v in enumerate(conteo.values):
        plt.text(conteo.index[j], v + 0.5, str(v), ha='center')

plt.tight_layout()
plt.show()

# Contar cuántos países tienen cada fuente como su principal emisor (ranking 1)
fuentes_principales = rankings[rankings == 1].count()
print("Número de países donde cada fuente es la principal emisora de CO2:")
print(fuentes_principales.sort_values(ascending=False))

# Determinar la fuente más contaminante en la mayoría de los países
fuente_mayoritaria = fuentes_principales.idxmax()
print(f"\nLa fuente más contaminante en la mayoría de países en 2007 fue: {fuente_mayoritaria}")
```

Esto sugiere que en 2007 la principal fuente de emisiones de CO₂ para la mayoría de países fueron los combustibles líquidos, lo cual es consistente con el uso extendido de derivados del petróleo en el transporte y la industria.
